```python
import pandas as pd
dataset=pd.read_csv("PreprocessedStudentsDataset.csv")
dataset
```

```python
dataset.isna().sum()
```

```python
#values=[value for value in range(startRange,endRange)]
values=[]
for value in range(40,60):
    values.append(value)
values
```

```python
#probabilities=[dist.pdf(value) for value in values]
from scipy.stats import norm
sample=dataset['ssc_p']
sample_mean=sample.mean()
sample_std=sample.std()
dist=norm(sample_mean,sample_std)
probabilities=[]
for value in values:
    probabilities.append(dist.pdf(value))
print(probabilities)
```
