```python
def stdNormalDistributionGraph(dataset):
    #import seaborn for plot the graph
    import seaborn as sns
    #calculate mean of the column
    mean=dataset.mean()
    #calculate std deviation of the column
    std=dataset.std()
    #store the coloumn value to variable values as list
    values=[i for i in dataset]
    #calculate the z score for each value
    z_score=[((j-mean)/std) for j in values]
    #Draw teh graph
    sns.distplot(z_score,kde=True)
    
```

```python
stdNormalDistributionGraph(dataset['ssc_p'])
```
